```python
# 8월 데이터만 선별해서 히스토그램으로 출력

import csv
f = open('../data/seoul_book.csv')
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        elif month == '01':
            jan.append(float(row[-1]))

import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(dpi = 300)
plt.hist(jan, bins = 100, color = 'b', label = '1월')
plt.hist(aug, bins = 100, color = 'r', label = '8월')
plt.legend()
plt.show()
```

```python
import csv
f = open('../data/seoul_book.csv')
data = csv.reader(f)
next(data)
month = [[], [], [], [], [], [], [], [], [], [], [], []]

for row in data:
    if row[-1] != '':
        month[int(row[0].split('-')[1]) - 1].append(float(row[-1])) # -1 하는 이유 >> 0 ~ 11 까지의 인덱스 값에 월별 데이터 저장

plt.boxplot(month)
plt.show()
```

```python
import csv
f = open('../data/seoul_book.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31):
    day.append([])
    
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11':
            day[int(row[0].split('-')[2]) - 1].append(float(row[-1]))
            
plt.boxplot(day)
plt.show()
```
