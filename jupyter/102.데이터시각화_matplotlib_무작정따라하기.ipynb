```python
import csv
import matplotlib.pyplot as plt
f = open('../data/seoul_book.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
print(result)
```

```python
import pandas as pd

df = pd.read_csv('../data/seoul.csv', encoding='CP949')
date = df.iloc[:, 2]
date_split = date.str[-2:]
obj = date_split[date_split == '11'].index.tolist()
data2 = df.iloc[obj, [2,5]].reset_index().drop(['index'], axis=1)
data2 = data2.set_index(data2['일시']).drop(['일시'], axis=1)

print(data2)
```
